```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
nf = "https://raw.githubusercontent.com/ygterl/EDA-Netflix-2020-in-R/master/netflix_titles.csv"
```

```python
df = pd.read_csv(nf) #reading the data
```

```python
df
```

```python
df.head()
```

```python
eda = df.drop(columns='description') #dropping the description column
```

```python
eda
```

```python
m = eda.dropna() #dropping rows with nan values
```

```python
m
```

# Separating TV Shows and Movies

```python
tv = m.loc[m['type']=='TV Show']
```

```python
tv
```

```python
movies = m.loc[m['type']=='Movie']
```

```python
movies
```

# Analysis of Movies and TV Shows on Netflix

```python
sns.countplot(x='rating',data=tv)
```

```python
sns.countplot(x='type',data=m)
```

There are more <b>movies</b> as compared to <b>TV shows</b> on <b>Netflix</b>

```python
fig, ax = plt.subplots(figsize=(15, 10))
sns.countplot(x='rating',data=movies)
```

# Top 10 Directors on Netflix

```python
tv['director'].value_counts()[:10].plot(kind='barh')
```

```python
movies['director'].value_counts()[:10].plot(kind='barh')
```

# Top Genres on Netflix

```python
tv['listed_in'].value_counts()[:5].plot(kind='pie')
```

```python
movies['listed_in'].value_counts()[:5].plot(kind='pie')
```

# Duration of movies and TV Shows on Netflix

```python
tv['duration'].value_counts()[:5].plot(kind='pie')
```

```python
movies['duration'].value_counts()[:10].plot(kind='pie')
```

```python
tv['release_year'].value_counts()[:10].plot(kind='pie')
```
